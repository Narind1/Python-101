```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('survey.csv')

# Display first few rows
print("Original Data:")
print(df.head())
print(f"\nShape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")

# Data Preprocessing
# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df['NAME'] = df['NAME'].fillna('Unknown')
df['EMAIL'] = df['Email address'].fillna('Unknown')
df['AGE'] = df['AGE'].fillna(df['AGE'].median())
df['SALARY'] = df['SALARY'].fillna(df['SALARY'].median())

# Derive EXPERIENCE column based on AGE and SALARY
# Experience logic: (Age - 22) * 0.8 + (Salary / 10000) * 0.2
# Assuming people start working at age 22
def calculate_experience(row):
    age = row['AGE']
    salary = row['SALARY']
    
    # Basic experience estimation
    potential_exp_from_age = max(0, age - 22)
    salary_factor = salary / 10000 if salary > 0 else 0
    
    # Weighted experience calculation
    experience = (potential_exp_from_age * 0.7) + (salary_factor * 0.3)
    return round(experience, 2)

# Apply the function to create EXPERIENCE column
df['EXPERIENCE'] = df.apply(calculate_experience, axis=1)

# Display processed data
print("\nProcessed Data:")
print(df.head())
print(f"\nNew columns: {df.columns.tolist()}")

# Save the processed data
df.to_csv('survey_processed.csv', index=False)
print("\nProcessed data saved to 'survey_processed.csv'")

# Display statistics
print("\nData Statistics:")
print(df[['AGE', 'SALARY', 'EXPERIENCE']].describe())
```
