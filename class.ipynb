```python
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def dog(self):
        return self.name,self.age
    def bark(self):
        print(f"{self.name} says woof")
dog1=Dog('a',1)
dog2=Dog('b',2)

print(dog1.name)
print(dog2.name)
dog1.dog()
dog2.bark()
```

The class should have the following:

Attributes:

owner (string) → account holder’s name

balance (float) → initial balance (default = 0)

Methods:

deposit(amount) → adds money to balance

withdraw(amount) → subtracts money if balance is enough, else print "Insufficient funds"

display() → prints owner’s name and current balance

```python
class Bank:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    def deposit(self,amount):
        return self.balance + amount
    def withdraw(self,amount):
        if amount > self.balance:
            return "insufficient balance"
        else:
            return self.balance - self.amount
    def display(self):
        print(f"{self.owner} is the owner and " f"{self.balance} is the current balance")

a = Bank('a',100)
b = Bank('b',200)
a.deposit(100)
```
