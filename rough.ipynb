```python
class Solution:
    def can_partition(self, num_str, target, index=0, current_sum=0):
        if index == len(num_str):
            return current_sum == target
        
        for j in range(index + 1, len(num_str) + 1):
            part = int(num_str[index:j])
            if self.can_partition(num_str, target, j, current_sum + part):
                return True
        
        return False

    def punishmentNumber(self, n):
        total = 0
        for i in range(1, n + 1):
            square_str = str(i * i)
            if self.can_partition(square_str, i):
                total += i * i
        return total


n = int(input())
print(Solution().punishmentNumber(n))
```

```python
class Solution:
    def romanToInt(self, s: str):
        roman_dict = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        num = 0
        prev_value = 0

        for char in reversed(s):  # Process from right to left
            curr_value = roman_dict[char]
            if curr_value < prev_value:  # If smaller value precedes larger, subtract it
                num -= curr_value
            else:
                num += curr_value
            prev_value = curr_value
        
        return num

# Example usage
solution = Solution()
result = solution.romanToInt("IV")
print(result)  # Output: 17
```

```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        i, j = 0, 0
        merged = []
        
        while i < len(nums1) and j < len(nums2):
            if nums1[i] < nums2[j]:
                merged.append(nums1[i])
                i += 1
            else:
                merged.append(nums2[j])
                j += 1

        merged.extend(nums1[i:])
        merged.extend(nums2[j:])

        n = len(merged)
        if n % 2 == 0:
            return (merged[n//2 - 1] + merged[n//2]) / 2  
        else:
            return merged[n//2]
sol = Solution()
sol.findMedianSortedArrays([1,2],[3,4])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Years
years = ["2024", "2030", "2070"]
x = np.arange(len(years))

# Central values (GW)
coal = [215, 240, 75]         # Coal capacity (approx, declining by 2070)
renewables = [220, 480, 1250] # Renewables/non-fossil (solar, wind, hydro)
others = [50, 90, 250]        # Nuclear, hydro, emerging clean

# Bar width
width = 0.6

# Stacked bar chart
fig, ax = plt.subplots(figsize=(10,6))

p1 = ax.bar(x, coal, width, label="Coal (GW)", color="grey")
p2 = ax.bar(x, renewables, width, bottom=coal, label="Renewables & Non-Fossil (GW)", color="green")
p3 = ax.bar(x, others, width, bottom=np.array(coal)+np.array(renewables), label="Others (GW)", color="blue")

# Add exact values on bars
for rects, data in zip([p1, p2, p3], [coal, renewables, others]):
    for rect, value in zip(rects, data):
        ax.text(rect.get_x() + rect.get_width()/2, rect.get_y() + rect.get_height()/2, 
                f"{value}", ha="center", va="center", fontsize=9, color="white", fontweight="bold")

# Labels and formatting
ax.set_ylabel("Installed Capacity (GW)")
ax.set_title("India’s Energy Mix Projection (2024–2070)", fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.legend(loc="upper left")
ax.set_ylim(0, 1700)
ax.grid(axis="y", alpha=0.3)

plt.tight_layout()
plt.show()
```
