```python
healthy_data = pd.read_csv('healthy.csv')
hypertension_data = pd.read_csv('hypertension.csv')

healthy_systolic = healthy_data['Systolic Blood Pressure(mmHg)'][:20].values
healthy_diastolic = healthy_data['Diastolic Blood Pressure(mmHg)'][:20].values

hypertension_systolic = hypertension_data['Systolic Blood Pressure(mmHg)'][:20].values
hypertension_diastolic = hypertension_data['Diastolic Blood Pressure(mmHg)'][:20].values

y = np.arange(1, 21) 
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(y, healthy_systolic, label="Healthy Systolic BP", marker='o', linestyle='-', color='blue')
plt.plot(y, hypertension_systolic, label="Hypertension Systolic BP", marker='s', linestyle='--', color='red')
plt.xlabel("Y Array (Index)")
plt.ylabel("Systolic Blood Pressure (mmHg)")
plt.title("Systolic Blood Pressure Comparison")
plt.legend()
plt.grid(True)
plt.xticks(ticks=y)
plt.yticks(ticks=np.arange(min(min(healthy_systolic), min(hypertension_systolic)),
                           max(max(healthy_systolic), max(hypertension_systolic)) + 1, 5)) 
plt.subplot(1, 2, 2)
plt.plot(y, healthy_diastolic, label="Healthy Diastolic BP", marker='o', linestyle='-', color='green')
plt.plot(y, hypertension_diastolic, label="Hypertension Diastolic BP", marker='s', linestyle='--', color='orange')
plt.xlabel("Y Array (Index)")
plt.ylabel("Diastolic Blood Pressure (mmHg)")
plt.title("Diastolic Blood Pressure Comparison")
plt.legend()
plt.grid(True)
plt.xticks(ticks=y) 
plt.yticks(ticks=np.arange(min(min(healthy_diastolic), min(hypertension_diastolic)),
                           max(max(healthy_diastolic), max(hypertension_diastolic)) + 1, 5))
plt.tight_layout()
plt.show()
```

```python

healthy_data = pd.read_csv('healthy.csv')

systolic = healthy_data['Systolic Blood Pressure(mmHg)'][:80]
diastolic = healthy_data['Diastolic Blood Pressure(mmHg)'][:80]

map_values = (systolic + 2 * diastolic) / 3

healthy_data['MAP'] = pd.Series(map_values, index=healthy_data.index[:80])

heart_rate = healthy_data['Heart Rate(b/m)'][:80]

correlation = np.corrcoef(map_values, heart_rate)[0, 1]
print(f"Correlation Coefficient between MAP and Heart Rate: {correlation:.3f}")

plt.figure(figsize=(10, 6))

plt.scatter(map_values, heart_rate, color='purple', alpha=0.6)

min_value = min(min(map_values), min(heart_rate))
max_value = max(max(map_values), max(heart_rate))
plt.plot([min_value, max_value], [min_value, max_value], color='red', linestyle='--', label='45-degree Line')

plt.xlabel('Mean Arterial Pressure (MAP) (mmHg)')
plt.ylabel('Heart Rate (b/m)')
plt.title('Scatter Plot: MAP vs Heart Rate with 45-degree Line')

plt.legend()
plt.show()
```

```python

healthy_data = pd.read_csv('healthy.csv')

systolic = healthy_data['Systolic Blood Pressure(mmHg)'][:80]
diastolic = healthy_data['Diastolic Blood Pressure(mmHg)'][:80]

map_values = (systolic + 2 * diastolic) / 3

healthy_data['MAP'] = pd.Series(map_values, index=healthy_data.index[:80])

heart_rate = healthy_data['Heart Rate(b/m)'][:80]

map_10_above = map_values * 1.1
map_10_below = map_values * 0.9
heart_rate_10_above = heart_rate * 1.1
heart_rate_10_below = heart_rate * 0.9

map_10_above_subset = healthy_data[map_values >= map_10_below]
map_10_below_subset = healthy_data[map_values <= map_10_above]
heart_rate_10_above_subset = healthy_data[heart_rate >= heart_rate_10_below]
heart_rate_10_below_subset = healthy_data[heart_rate <= heart_rate_10_above]

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.scatter(map_10_above_subset['MAP'], heart_rate_10_above_subset['Heart Rate(b/m)'], color='blue', alpha=0.6)
plt.title('MAP 10% Above vs Heart Rate 10% Above')
plt.xlabel('MAP (10% Above)')
plt.ylabel('Heart Rate (10% Above)')
plt.grid(True)

min_val = min(map_10_above_subset['MAP'].min(), heart_rate_10_above_subset['Heart Rate(b/m)'].min())
max_val = max(map_10_above_subset['MAP'].max(), heart_rate_10_above_subset['Heart Rate(b/m)'].max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='45-degree Line')

plt.plot([min_val, max_val], [min_val + 10, max_val + 10], color='green', linestyle='-.', label='45-degree Line + 10 Units')

plt.plot([min_val, max_val], [min_val - 10, max_val - 10], color='orange', linestyle='-.', label='45-degree Line - 10 Units')

plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(map_10_below_subset['MAP'], heart_rate_10_below_subset['Heart Rate(b/m)'], color='red', alpha=0.6)
plt.title('MAP 10% Below vs Heart Rate 10% Below')
plt.xlabel('MAP (10% Below)')
plt.ylabel('Heart Rate (10% Below)')
plt.grid(True)

min_val = min(map_10_below_subset['MAP'].min(), heart_rate_10_below_subset['Heart Rate(b/m)'].min())
max_val = max(map_10_below_subset['MAP'].max(), heart_rate_10_below_subset['Heart Rate(b/m)'].max())
plt.plot([min_val, max_val], [min_val, max_val], color='blue', linestyle='--', label='45-degree Line')

plt.plot([min_val, max_val], [min_val + 10, max_val + 10], color='green', linestyle='-.', label='45-degree Line + 10 Units')

plt.plot([min_val, max_val], [min_val - 10, max_val - 10], color='orange', linestyle='-.', label='45-degree Line - 10 Units')

plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

files = ['2_1.csv', '2_2.csv', '2_3.csv']
all_data = []

for file in files:
    data = pd.read_csv(file, header=None)  
    values = data.values.flatten() 
    numeric_values = pd.to_numeric(values, errors='coerce')  
    cleaned_values = pd.Series(numeric_values).dropna()  
    all_data.extend(cleaned_values)  

y = pd.Series(all_data)

x = range(len(y)) 

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='', color='blue', label='Combined Data') 

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Combined Data from Multiple Files')
plt.legend()

plt.grid(True) 
```
