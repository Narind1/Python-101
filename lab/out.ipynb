```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

```python
np.random.seed(42)
data = np.concatenate([np.random.normal(50,10,100),np.random.normal(100,20,50),np.random.normal(200,50,10)])
data2 = np.concatenate([np.random.normal(50,10,100),np.random.normal(50,20,50),np.random.normal(50,50,10)])
df2 = pd.DataFrame(data2,columns = ["X"])

df = pd.DataFrame(data,columns = ["X"])
df.rename(columns={df.columns[0]: "X"}, inplace=True)
print(df.head())
```

```python
data1 = data[::-1]
df1 = pd.DataFrame(data1, columns=["X"])
print(df1.head())

plt.figure(figsize=(10,6))
sns.boxplot(x=df['X'])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df1['X'])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df2['X'])
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(df['X'], kde=True, color='red', label='df', alpha=0.5)
sns.histplot(df1['X'], kde=True, color='blue', label='df1', alpha=0.5)

plt.title('Data Distribution Comparison')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness = skew(df['X'])
skewness1 = skew(df1['X'])


print(f"Skewness of the data: {skewness:.2f}")
print(f"Skewness of the 2nd data: {skewness1:.2f}")
```

```python
std_dev = df ['X']. std ()
print (f" Standard Deviation : { std_dev :.2f}")

Q1 = df ['X']. quantile (0.25)
Q3 = df ['X']. quantile (0.75)
IQR = Q3 - Q1
print (f" Interquartile Range ( IQR ): { IQR :.2f}")
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df [( df ['X'] < lower_bound ) | ( df ['X'] >
upper_bound )]
print (" Outliers detected")
print(outliers)
```
