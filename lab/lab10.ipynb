```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Housing.csv')

plt.figure(figsize=(8, 4))
sns.histplot(data['price'], kde=True, bins=20, color='blue', edgecolor='black')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(x='furnishingstatus', y='price', data=data, palette='coolwarm')
plt.title('Boxplot: Furnishing Status vs Price')
plt.xlabel('Furnishing Status')
plt.ylabel('Price')
plt.show()
```
