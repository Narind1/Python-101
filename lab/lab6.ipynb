```python
import pandas as pd

data = {
    "State": ["Karnataka", "Maharashtra", None, "West Bengal", "Gujarat"],
    "Crop": ["Rice", None, "Rice", "Wheat", "Rice"],
    "Area": [1000, 1500, None, 1300, 1100],
    "Yield": [2500, None, 2700, 2900, None],
    "Year": [2021, 2021, 2021, 2021, 2022]
}

df = pd.DataFrame(data)
a=df.copy()

df_dropped = df.dropna()
print("\nDataFrame after dropping rows with missing values:\n", df_dropped)

median_area = df['Area'].median()
median_yield = df['Yield'].median()

a['Area'] = a['Area'].fillna(median_area)
a['Yield'] = a['Yield'].fillna(median_yield)

a['State'] = a['State'].fillna(a['State'].mode()[0])
a['Crop'] = a['Crop'].fillna(a['Crop'].mode()[0])

print("\nDataFrame after replacing missing values with median and mode:\n", a)

df_ffill = df.fillna(method='ffill')
print("\nDataFrame after forward filling:\n", df_ffill)

df_bfill = df.fillna(method='bfill')
print("\nDataFrame after backward filling:\n", df_bfill)

print("\nAccess first row with iloc:\n", df.iloc[0])
print("\nAccess rows 1 to 3 and columns 2 and 3 with iloc:\n", df.iloc[1:4, 2:4])
print("\nValue at 2nd row and 3rd column with iloc:\n", df.iloc[1, 2])


print("\nAccess rows 1 to 3 with loc:\n", df.loc[1:3])
df.loc[(df.State=='Karnataka'),df.Year]=2028
print("\nAccess rows where 'State' is 'Karnataka' with loc:\n", df.loc[df['State'] == 'Karnataka'])
```
