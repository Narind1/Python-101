```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("car.csv", encoding='ISO-8859-1')

X = data['annual Salary'].values
Y = data['car purchase amount'].values

mean_X = np.mean(X)
mean_Y = np.mean(Y)

n = len(X)
numerator = np.sum((X - mean_X) * (Y - mean_Y))
denominator = np.sum((X - mean_X) ** 2)
b1 = numerator / denominator
b0 = mean_Y - (b1 * mean_X)

print(f"Intercept (b0): {b0}")
print(f"Slope (b1): {b1}")

def predict(x):
    return b0 + b1 * x

predicted_Y = predict(X)

plt.scatter(X, Y, color="blue", label="Data Points")
plt.plot(X, predicted_Y, color="red", label="Regression Line")
plt.xlabel("Annual Salary")
plt.ylabel("Car Purchase Amount")
plt.legend()
plt.title("Linear Regression of Car Purchase Amount vs Annual Salary")
plt.show()
```
