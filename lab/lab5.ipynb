```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'winequality-red.csv'
wine_data = pd.read_csv(file_path)

print(wine_data.describe())

corr_matrix = wine_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

wine_data.hist(bins=15, figsize=(15, 10), layout=(4, 3), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Wine Quality Features')
plt.show()

plt.figure(figsize=(15, 10))
for i, column in enumerate(wine_data.columns[:-1], 1):
    plt.subplot(4, 3, i)
    sns.boxplot(x='quality', y=column, data=wine_data)
    plt.title(f'Boxplot of {column} by Quality')
plt.tight_layout()
plt.show()

sns.pairplot(wine_data, hue='quality', palette='coolwarm')
plt.suptitle('Pairplot of Wine Quality Data', y=1.02)
plt.show()
```
