```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler,MaxAbsScaler,LabelEncoder,OneHotEncoder,OrdinalEncoder
```

```python
np.random.seed(42)
data = np.random.randint(0,100,size = (100,4))
df = pd.DataFrame(data,columns = ['Feature1','Feature2','Feature3','Feature4'])
data1 = np.random.randint(0,1000,size = (1000,1))
df1 = pd.DataFrame(data1,columns = ['Feature5'])
df2 = pd.concat([df, df1], axis=1)
df2.head()
```

```python
min_max_scaler = MinMaxScaler()
df_min_max_scaled = pd.DataFrame(min_max_scaler.fit_transform(df2),columns = df2.columns)
df_min_max_scaled.head()
```

```python
standard_scaler = StandardScaler()
df_standard_scaled = pd.DataFrame(standard_scaler.fit_transform(df2),columns = df2.columns)
df_min_max_scaled.head()
```

```python
robust_scaler = RobustScaler()
df_standard_scaled = pd.DataFrame(robust_scaler.fit_transform(df2),columns = df2.columns)
df_standard_scaled.head()
```

```python
d = {
 'Color' : [ 'Red ' , ' Blue ' , ' Green ' , ' Blue ' , 'Red ' , ' Green ' , 'Black'] ,
 'Size' : [ 'S' , 'M' , 'L' , 'M' , 'S' , 'L' , 'XL'] ,
 'Price' : [ 10 , 20 , 30 , 20 , 10 , 30, 10]
 }
df = pd . DataFrame ( d )
```

```python
label_encoder = LabelEncoder()
df['Color_LabelEncoded']= label_encoder.fit_transform(df['Color'])
df.head()
```

```python
one_hot_encoder = OneHotEncoder(sparse_output = False)
one_hot_encoded = one_hot_encoder.fit_transform(df[['Color']])
df_one_hot_encoded = pd.DataFrame(one_hot_encoded,columns = one_hot_encoder.get_feature_names_out(['Color']))
df_one_hot_encoded.head()
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['S', 'M', 'L','XL']])
df['Size_OrdinalEncoded'] = ordinal_encoder.fit_transform(df[['Size']])
df.head()
```
